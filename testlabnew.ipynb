```python
import numpy as np
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense
```

```python

training_data = np.array([[2,3,1],[1,2,2],[2,2,2],[5,3,1] ,[6,2,2] , [8,2,2]], "float32")

target_data = np.array([[1],[1],[1],[0],[0],[0]], "float32")
```

```python
test_data = np.array([[7,3,1] , [3,2,2]] , "float32")
test_value=np.array([[0],[1]] , "float32")
```

```python
model = Sequential()
model.add(Dense(3, input_dim=3, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='mean_squared_error',
              optimizer='adam',
              metrics=['binary_accuracy'])
```

```python
history = model.fit(training_data, target_data, epochs=500, verbose=2 , batch_size=2)
```

```python
history.history
```

```python
print( model.predict(training_data).round())
```

```python
res= model.evaluate(test_data,test_value)
```
